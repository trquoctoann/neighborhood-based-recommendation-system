<h1><center>Association Rule</center></h1>

```python
import pandas as pd 
import numpy as np
import os 
import plotly.graph_objects as go
import plotly.express as px
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
path = "D:\Giáo Trình\Kỳ 5\Kho và Khai Phá Dữ Liệu\Project\Processed Data"
catg = pd.read_csv(os.path.join(path, 'transaction_data.csv'))
catg.head(5)
```

## Category's Association Rules

```python
catg.columns = catg.columns.str.replace('catg_', '')
catg = catg.drop(columns = 'order_purchase_timestamp')
catg.head(5)
```

```python
catg = catg.groupby('customer_unique_id').sum().reset_index()
for i in catg.columns[1:]:
    catg[i] = catg[i].apply(lambda x : 1 if x > 0 else 0)
catg.head(5)
```

```python
frequent_catgsets = fpgrowth(catg.iloc[:,1:], min_support = 0.00002, use_colnames = True)
frequent_catgsets['length'] = frequent_catgsets['itemsets'].apply(lambda x : len(x))
frequent_catgsets
```

```python
catg_rules = association_rules(frequent_catgsets, metric = 'lift', min_threshold = 2)
catg_rules.sort_values('confidence', ascending = False)
```

```python
catg_rules.to_csv(path + '\catg_rules.csv', index = False)
```

## Product's Association Rules

```python
prod = pd.read_csv(os.path.join(path, 'order_data.csv'))
prod.head(5)
```

```python
def get_prod_list(row):
    tempList = []
    for prod in row.unique() : 
        tempList.append(prod)
    return tempList
prod_list = prod[['order_id', 'product_id']].groupby('order_id').agg(get_prod_list).reset_index().product_id
prod_list = list(prod_list)
prod_list[:5]
```

```python
tEncoder = TransactionEncoder()
trans_array = tEncoder.fit(prod_list).transform(prod_list)
trans_prod = pd.DataFrame(trans_array, columns = tEncoder.columns_)
trans_prod.head(5)
```

```python
frequent_prodsets = fpgrowth(trans_prod, min_support = 0.00005, use_colnames = True)
frequent_prodsets['length'] = frequent_prodsets['itemsets'].apply(lambda x : len(x))
frequent_prodsets
```

```python
prod_rules = association_rules(frequent_prodsets, metric = 'confidence', min_threshold = 0.4)
prod_rules.sort_values('confidence', ascending = False)
```

```python
prod_rules.to_csv(path + '\prod_rules.csv', index = False)
```
