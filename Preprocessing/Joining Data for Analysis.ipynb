                                                   Joining Data for Analysis

```python
import pandas as pd 
import numpy as np
import os
```

### Customer Data

```python
path = "D:\Giáo Trình\Kỳ 5\Kho và Khai Phá Dữ Liệu\Project"
customerData = pd.read_csv(os.path.join(path,'Data\olist_customers_dataset.csv'))
customerData.head()
```

```python
locationData = pd.read_csv(os.path.join(path, 'Data\olist_geolocation_dataset.csv'))
locationData.head()
```

```python
iloc_location = locationData.iloc[:, :3].groupby('geolocation_zip_code_prefix').mean().reset_index()
colsCustomer = ['customer_id', 'customer_unique_id', 'customer_zip_code_prefix'
            ,'customer_city', 'customer_state', 'geolocation_lat', 'geolocation_lng']
customerAfter = pd.merge(left = customerData, right = iloc_location, left_on = 'customer_zip_code_prefix'
                        ,right_on = 'geolocation_zip_code_prefix', how = 'left')[colsCustomer]
customerAfter.head()
```

```python
customerAfter.to_csv(path + '\Processed Data\customer_data.csv', index = False)
```

### Order Data

```python
order_items = pd.read_csv(os.path.join(path, 'Data\olist_order_items_dataset.csv'))
order_items.head()
```

```python
order_status = pd.read_csv(os.path.join(path, 'Data\olist_orders_dataset.csv'))
order_status.head()
```

```python
paymentData = pd.read_csv(os.path.join(path, 'Data\olist_order_payments_dataset.csv'))
paymentData.head()
```

```python
order_payment = paymentData[['order_id', 'payment_installments', 'payment_value']].groupby('order_id').sum().reset_index()
order_payment.head()
```

```python
colsJoin1 = ['order_id', 'order_item_id', 'customer_id', 'product_id', 'seller_id', 'order_status', 'order_purchase_timestamp'
        , 'order_delivered_customer_date', 'order_estimated_delivery_date', 'price', 'freight_value']
orderJoin1 = pd.merge(left = order_items, right = order_status, on = 'order_id', how = 'inner')[colsJoin1]
orderJoin1.head()
```

```python
colsJoin2 = ['order_id', 'order_item_id', 'customer_unique_id', 'product_id', 'seller_id', 'order_status', 'order_purchase_timestamp'
            , 'order_delivered_customer_date', 'order_estimated_delivery_date', 'price', 'freight_value']
orderJoin2 = pd.merge(left = orderJoin1, right = customerData, on = 'customer_id', how = 'inner')[colsJoin2]
orderJoin2.head()
```

```python
cols_need = ['order_id', 'order_item_id', 'customer_unique_id', 'product_id', 'seller_id', 'order_status', 'order_purchase_timestamp'
            , 'order_delivered_customer_date', 'order_estimated_delivery_date', 'price', 'freight_value', 'payment_value']
orderAfter = pd.merge(left = orderJoin2, right = order_payment, on = 'order_id', how = 'left')[cols_need]
orderAfter.head()
```

```python
orderAfter.to_csv(path + '\Processed Data\order_data.csv', index = False)
```

### Product Data

```python
productData = pd.read_csv(os.path.join(path, 'Data\olist_products_dataset.csv'))
productData.head()
```

```python
transData = pd.read_csv(os.path.join(path, 'Data\product_category_name_translation.csv'))
transData.head()
```

```python
productMerge = pd.merge(left = productData, right = transData, on = 'product_category_name', how = 'left')
productAfter = productMerge.drop(['product_category_name'], axis = 1)
productAfter.head()
```

```python
productAfter.to_csv(path + '\Processed Data\product_data.csv', index = False)
```

### Transaction Data

```python
transRaw = pd.merge(left = orderAfter, right = productAfter, on = 'product_id', how = 'inner')[['customer_unique_id'
                    , 'order_purchase_timestamp', 'product_category_name_english']]
transaction_data = pd.get_dummies(data= transRaw, columns= ['product_category_name_english'], prefix = 'catg').groupby([
                    'customer_unique_id','order_purchase_timestamp']).sum().reset_index()
transaction_data.head()
```

```python
transaction_data.to_csv(path + '\Processed Data\\transaction_data.csv', index = False)
```

### Seller Data

```python
sellerData = pd.read_csv(os.path.join(path, 'Data\olist_sellers_dataset.csv'))
sellerData.head()
```

```python
seller_loc = pd.merge(left = sellerData,right = iloc_location,left_on = 'seller_zip_code_prefix'
                    , right_on= 'geolocation_zip_code_prefix', how = 'left')
seller_loc.rename(columns={'geolocation_lat':'seller_lat','geolocation_lng':'seller_lng'}, inplace= True)
seller_loc.head()
```

```python
customerAfter.rename(columns={'geolocation_lat':'customer_lat','geolocation_lng':'customer_lng'}, inplace= True)
customerAfter.head()
```

```python
cols= ['order_id','customer_id','order_purchase_timestamp','order_delivered_customer_date','order_estimated_delivery_date','seller_id']
order_seller = pd.merge(left= order_items,right= order_status,on= 'order_id',how= 'left')[cols]
order_seller.head(5)
```

```python
cols= ['customer_id','customer_unique_id','customer_lat','customer_lng','customer_city']
cust_seller = pd.merge(left= order_seller,right= customerAfter[cols],on= 'customer_id',how= 'left')
cust_seller.head(5)
```

```python
cols_need = ['order_id','customer_unique_id','seller_id','order_purchase_timestamp','order_delivered_customer_date'
            , 'order_estimated_delivery_date', 'seller_city', 'customer_city', 'seller_lat', 'seller_lng', 'customer_lat', 'customer_lng']
colsSeller= ['seller_id','seller_city','seller_lat','seller_lng']
delivery_data = pd.merge(left= cust_seller,right= seller_loc[colsSeller],on= 'seller_id',how= 'left')[cols_need]
delivery_data.head()
```

```python
delivery_data.to_csv(path + '\Processed Data\delivery_data.csv', index = False)
```

### Product Review Score Data

```python
reviewScoreData = pd.read_csv(os.path.join(path, 'Data\olist_order_reviews_dataset.csv'))
reviewScoreData.head()
```

```python
colsJoin1 = ['product_id', 'review_score']
reviewScoreJoin1 = pd.merge(left = reviewScoreData, right = orderAfter, on = 'order_id', how = 'inner')[colsJoin1]
reviewScoreJoin1.head()
```

```python
colsReview = ['product_id', 'product_category_name_english', 'review_score']
reviewScoreAfter = pd.merge(left = productAfter, right = reviewScoreJoin1, on = 'product_id', how = 'inner')[colsReview].groupby([
                        'product_category_name_english', 'product_id']).mean().reset_index()

reviewScoreAfter.head()
```

```python
reviewScoreAfter.to_csv(path + '\Processed Data\product_review_score.csv', index = False)
```
